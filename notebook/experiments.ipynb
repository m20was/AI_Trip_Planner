```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

# 1. Initialize your environment keys
load_dotenv()

# 2. Instantiate the ChatGroq client model cleanly
llm = ChatGroq(model="openai/gpt-oss-120b")

# 3. Invoke a response
response = llm.invoke("hi")
print(response.content)
```

```python
from langchain.tools import tool
@tool # This is a tool decorator
def multiply(a: int, b: int) -> int:
    """
    Multiply two integers.

    Args:
        a (int): The first integer.
        b (int): The second integer.

    Returns:
        int: The product of a and b.
    """
    return a * b
```

```python
from typing import ClassVar, Type
from pydantic import BaseModel, Field
from langchain_core.tools import StructuredTool

# 1. Define the input validation schema first
class WeatherInput(BaseModel):
    city: str = Field(..., description="City name")
    units: str = Field("metric", description="metric or imperial")

# 2. Define the core python function next
def get_weather(city: str, units: str = "metric") -> str:
    """
    Get the weather for a given city.
    """
    return f"The weather in {city} is sunny using {units} units."

# 3. Create the tool instance (Standard Approach)
weather_tool = StructuredTool.from_function(
    func=get_weather,
    name="get_weather",
    description="Fetches real-time weather data for a city",
    args_schema=WeatherInput,   # This is a pydantic class, ab ye model work karega
)
```
